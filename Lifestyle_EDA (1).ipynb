Eksploratory Data Analysis : Life Style Dataset

Eksploratory ini bertujuan untuk memahami pola olahraga, kalori yang terbakar dengan jenis olahraganya berdasarkan usia, dan jenis kelamin

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Dataset Overview:
Dataset ini berisikan fitur fitur utama yang dipilih sebagai berikut:
  1. Age : usia peserta
  2. Gender : jenis kelamin peserta (Male/Female)
  3. Calories Burned : kalori yang terbakar saat workout dalam satu jam
  4. Workout_Type : Jenis olahraga yang dilakukan

  pada langkah ini juga termasuk kedalam proses melihat struktur dataset dan mengubah tipe data tiap kolom supaya sesuai dengan pengolahan data

```python
df = pd.read_csv('/content/drive/MyDrive/Portofolio_Data/Final_data.csv')
df = df[['Age','Gender','Calories_Burned', 'Workout_Type']]
df
```

```python
df.info()
```

```python
#mengubah tipe data tiap kolom
df['Age'] = df['Age'].astype(int)
df['Gender'] = df['Gender'].astype('category')
df['Calories_Burned'] =df['Calories_Burned'].astype(int)
df['Workout_Type']= df['Workout_Type'].astype('category')
df
```

```python
#melihat apakah sudah berubah tipe data tiap kolom
df.info()
```

1. Distribusi Usia

Hasil analisis pada kolom usia menunjukan bahwa rata-rata usia peserta berkisar antara usia 38 tahun, hal ini berarti mayoritas peserta berada di usia dewasa yang cukup produktif untuk menjaga kebugaran tubuh dengan berolahraga

```python
#Melihat distribusi usia peserta
df['Age'].describe()
```

Rata-Rata usia pada tiap jenis olahraga

Hasil analisa menunjukan bahwa pada keempat jenis olahraga tersebut baik cardio, HIIT, Strength Training dan juga Yoga, menujukan rata-rata usia sekiatar 38 tahun. hal ini sesuai dengan distribusi kolom usia pada dataset yang menunjukan hal yang sama, bahwa rata-rata usia peserta yag sudah berusia dewasa pun tetap menjaga kesehatan jasmani mereka dengan terus berolahraga.

```python
#mengetahui rata-rata usia pada tiap jenis workout
df.groupby('Workout_Type')['Age'].mean()
```

2. Melihat distribusi jenis kelamin peserta

Berdasarkan hasil analisis menunjukan bahwa jumlah peserta wanita lebih banyak dibandingkan dengan jumlah peserta pria

```python
#Menghitung jumlah peserta lebih banyak wanita atau laki-laki
df['Gender'].value_counts()
```

3. Jenis workout pada tiap jenis kelamin.

Hasil analisis menunjukan bahwa, perempuan lebih menggemari olah raga dengan tipe strength sedangkan laki-laki lebih memilih olahraga cardio.
Pola tersebut menunjukan bahwa perempuan cenderung meningkatkan kekuatan dan massa otot, sedangkan laki-laki cenderung memilih cardio untuk meningkatkan kebugaran jantung dan paru-paru

```python
#mengetahui jenis workout yang digunakan tiap jenis kelamin
df.groupby('Gender')['Workout_Type'].value_counts()
```

```python
#Visualisasi
Countdata = df.groupby(['Gender','Workout_Type']).size().reset_index(name='Count')
plt.figure(figsize=(10,5))
sns. barplot(
    data=Countdata,
    x='Workout_Type',
    y='Count',
    hue='Gender'
)
plt.title('Jenis Workout yang digunakan tiap Gender')
plt.show()
```

4. Rata-rata kalori terbakar pada masing-masing jenis workout berdasarkan jenis kelamin.

Hasil analisa menunjukan bahwa jenis olahraga HIIT merupakan jenis olahraga yang membakar kalori tertinggi, sekitar 1658 kalori per jam baik pada perempuan atau laki-laki, selanjutnya adalah strength training, cardio, dan yang paling rendah adalah yoga, yaitu membakar sekitar 896 kalori per jam.

```python
#untuk mengetahui rata-rata kalori terbakar tiap gender dan jenis workoutnya
df.groupby(['Gender','Workout_Type'])['Calories_Burned'].mean().reset_index().sort_values('Calories_Burned', ascending=False)
```

```python
#visualisasi rata-rata kalori terbakar tiap gender
plt.figure(figsize=(10,5))
sns.barplot(
    data=df,
    x='Workout_Type',
    y='Calories_Burned',
    hue='Gender',
    estimator= 'mean'
)
plt.title('Rata-rata Kalori Terbakar tiap Gender')
plt.show()
```

Kesimpulan :

Secara Keseluruhan:
1. Rata - Rata Peserta pada masing-masing jenis olahraga: 38 Tahun
2. Peserta perempuan lebih banyak dibandingkan laki-laki
3. Peserta perempuanlebih menyukai olahraga strength training sedagkan laki-laki lebih memilih cardio
4. Jenis olahraga HIIT merupakan jenis olahraga yang membakar kalori paling banyak baik pada laki=laki maupun perempuan
